```python
df_oct=df_unit_sales.loc['2020-10-01':'2020-10-30']
```

```python
df_nov=df_unit_sales.loc['2020-11-01':'2020-11-30']
df_nov
```

```python
df_dec=df_unit_sales.loc['2020-12-01':'2020-12-30']
df_dec.shape
```

```python
sns.set(rc={'figure.figsize':(18,8)})

fig,axes=plt.subplots(1,3)

bp_oct=sns.boxplot(x='Sales',data=df_oct,ax=axes[0],color='r')
bp_nov=sns.boxplot(x='Sales',data=df_nov,ax=axes[1],color='b')
bp_dec=sns.boxplot(x='Sales',data=df_dec,ax=axes[2],color='g')

bp_oct.set(title='October Sales')
bp_nov.set(title='November Sales')
bp_dec.set(title='December Sales')

plt.show()
```

## From the above figure we can see that the sales during the month of october has been consistent and didnt have any outliers but sales during november and december has outliers which shows drastic drop and spike in sales.

### November sees a dip in the beginning of the month and by the end of the month there is a spike .December has recorded the highest sales .

```python
sns.set(rc={'figure.figsize':(18,8)})
sns.lineplot(x='Date',y='Sales',data=sales_df)
plt.show()
```

## The line plot suugests the same as the boxplot. Oct has large and consistent sales throughout the month.November sees a dip initially and towards the end , there is a spike. December records the highest sales and has a spike beginning of the month.
