```python
import uuid
print("Welcome to the demo market place")
sample_database={
    'user1':{'password':'user123'},
    'admin12':{'password':'admin123'}
}
categories = {
    1: "Snacks",
    2: "Clothing",
    3: "Electronics"
}
user_cart={}
products_db = {
    101: {"name": "Diary Milk (Silk)", "category_id": 1, "price": 250},
    102: {"name": "Shirt", "category_id": 2, "price": 2000},
    103: {"name": "Pant", "category_id": 2, "price": 1500},
    104: {"name": "PS5", "category_id": 3, "price": 55500}
}

def display_user_menu():
    print("""
    Options
    
1. View catalog
2. Add to cart
3. Remove from cart
4. Checkout
5. Exit
 """)
    
def display_admin_menu():
    print("""
    Options
    
1. View catalog
2. Add new item to catalog
3. Remove item from catalog
4. Add category
5. Remove category
6. Exit
 """)


def add_to_cart(session_id):
    product_id=int(input('enter the product id'))
    quantity= int(input('enter the quantity'))
    if session_id not in user_cart:
        user_cart[session_id] = {}
    
    if product_id in products_db:
        if product_id in user_cart[session_id]:
            user_cart[session_id][product_id] += quantity
        else:
            user_cart[session_id][product_id] = quantity
        print(f"Added {quantity} of {products_db[product_id]['name']} to your cart.")
    else:
        print("Invalid product ID.")

def delete_from_cart(session_id):
    product_id=int(input("enter then product id to remove"))
    if session_id in user_cart and product_id in user_cart[session_id]:
        del user_cart[session_id][product_id]
        print(f"Removed product {products_db[product_id]['name']} from your cart.")
    else:
        print("Product not in your cart.")
    
def add_item_to_catalog():
    name=input('enter the name')
    categoryid=int(input('enter the category id'))
    price=int(input('enter the price'))
    new_product_id = max(products_db.keys()) + 1
    products_db[new_product_id] = {"name": name, "category_id": categoryid, "price": price}
    print(f"Product {name} added successfully with ID {new_product_id}.")

def remove_product_admin():
    product_id=int(input('enter the product id'))
    if product_id in products_db:
        del products_db[product_id]
        print(f"Product {product_id} removed successfully.")
    else:
        print(f"Product with ID {product_id} not found.")
def checkout(session_id):
    if session_id not in user_cart or not user_cart[session_id]:
        print("Your cart is empty.")
        return
    
    total_price = 0
    for product_id, quantity in user_cart[session_id].items():
        total_price += products_db[product_id]["price"] * quantity
    
    print(f"Total Price: Rs. {total_price}")
    print("""
    Payment Options
1.  UPI
2.  Paypal
3.  Net banking
    """)
    choice=int(input("enter your payment choice"))
    if choice == 1:
        print(" \n Proceeding with UPI")
    elif choice ==2:
        print(" \n Redirecting to Paypal")
    elif choice ==3:
        print("Redirecting to bank portal")
    else:
        print("wrong choice")
    print("Your payment was successful and Order has been placed")

def add_category_admin():
    name=input('enter the name of the new category')
    new_category_id = max(categories.keys()) + 1
    categories[new_category_id] = name
    print(f"Category '{name}' added with ID {new_category_id}.")


def remove_category_admin():
    category_id=int(input('enter the category id to remove'))
    if category_id in categories:
        del categories[category_id]
        print(f"Category {category_id} removed successfully.")
    else:
        print(f"Category with ID {category_id} not found.")
    
def display_products():
    print("\nProduct Catalog:")
    for product_id, details in products_db.items():
        category_name = categories[details["category_id"]]
        print(f"{product_id}: {details['name']} | Category: {category_name} | Price: {details['price']}")
        
def start_user_session(username):
    return str(uuid.uuid4())

def authenticate(username,password):
    if username in sample_database and sample_database[username]['password']==password:
        return True
    return False

def start():   
    while True:
        user_input=input("\nAre you a user/admin")
        if user_input=='user':
            username=input('enter your username')
            password=input('enter the password')
            print("\n")
            if authenticate(username,password):
                session_id=start_user_session(username)
                print(f"Welcome {username}, Your session id : {session_id}", end="\n")
                while True: 
                    display_user_menu()
                    choice=int(input("Enter your choice"))
                    if choice ==1:
                        display_products()
                    elif choice ==2:
                        add_to_cart(session_id)
                    elif choice ==3:
                        delete_from_cart(session_id)
                    elif choice ==4:
                        checkout(session_id)
                    elif choice ==5:
                        print('\n Thank you')
                        return False
                        break
                    else:
                        print("wrong choice")
            else:
                print('Invalid')

        
        elif user_input=='admin':
            admin_cred=input('enter the admin credentials')
            admin_pass=input('enter the password')
            if authenticate(admin_cred,admin_pass):
                session_id=start_user_session(admin_cred)
                print(f'welcome {admin_cred}, session id : {session_id}')
                while True:
                    display_admin_menu()
                    choice=int(input("Enter your choice"))
                    if choice == 1:
                        display_products()
                    elif choice ==2:
                        add_item_to_catalog()
                    elif choice ==3:
                        remove_product_admin()
                    elif choice ==4:
                        add_category_admin()
                    elif choice ==5:
                        remove_category_admin()
                    elif choice ==6:
                        print(" \n Thank you")
                        return False
                        break
                    else:
                        print("Wrong choice")
                        
        else:
            print('Invalid user type')
            continue
            
        
        
    
if __name__ == "__main__":
    start()
```
